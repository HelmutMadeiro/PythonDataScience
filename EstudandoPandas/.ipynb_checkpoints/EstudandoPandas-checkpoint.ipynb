```python
#importando pandas
import pandas as pd

#importando um arquivo xlsx 
dados = pd.read_excel('Vendas.xlsx')
dadosVendas = pd.DataFrame (dados)
```

```python
#alguns meta dados sobre o DataFrame 
#buscando o total de linhas e colunas
dadosVendas.shape
#pengando o tipo de cada coluna 
dadosVendas.dtypes
#buscando uma lista com os nomes de cada coluna
dadosVendas.columns.to_list()
```

```python

#renomando as colunas 
dadosVendas = dadosVendas.rename(columns={   'Código Venda':'CodigoVenda',
                                             'Data':'Data',
                                             'ID Loja':'IDLoja',
                                             'Produto':'Produto',
                                             'Quantidade':'Quantidade',
                                             'Valor Unitário':'ValorUnitario',
                                             'Valor Final':'ValorFinal'})

#mostrar apenas as 3 primeiras linhas 
dadosVendas.head(3)

```

```python
#podemos ter algumas coisas importantes com esse metodo 
# ele vai em cada coluna que são calculaveis ( numeros e tras algumas medições)
dadosVendas.describe()
```

```python
#pegando apenas uma coluna de um DataFrame ( Series)
#um DataFrame é composto de 2 ou mais de Series 
#uma Series é uma coluna de  um DataFrame

produtos = dadosVendas['Produto']
produtos
```

```python
#Pegando duas colunas de um DataFrame
produtos = dadosVendas[['Produto','IDLoja']]
produtos
```

```python
#usando o loc 
#pengando apenas uma linha do DataFrama ( lembrando que agora retorna apenas uma Siries)
#lembrando que o loc viaja usando o index da sua tabela  (DataFrame)
dadosVendas.loc[1]
```

```python
#usando o loc para buscar mais que uma linha

dadosVendas.loc[1:5]
```

```python
#pegar linhas que correspodem a uma condição
#dentro do loc passamos o DataFrame e a coluna que queiramos fazer a condição 
#porem temos aqui todas as colunas 
dadosVendas.loc[dadosVendas['IDLoja'] =='Norte Shopping']
```

```python
#bucando uma condição e trazendo apenas colunas especificas
# vendo o padrão do loc DataFrame.loc[linhas,colunas]
# a condição trará todas as linhas que se encaixam da sua logica, após a virgula vc colocará as colunas que vc quer ver
dadosVendas.loc[dadosVendas['IDLoja'] =='Norte Shopping', #linhas obdecendo a condição 
                           ['IDLoja','Produto','Quantidade'] #colunas selecionadas 
               ]

```

```python
#bucar valor especifico

dadosVendas.loc[1,'Produto']
```

```python
#adcionar colunas 

# a patie de uma coluna que existe
dadosVendas['Comissao'] = dadosVendas['ValorFinal'] * 0.05

dadosVendas.head(2)
```

```python
# Criando colunas com valor padrão
#podemos fazer assim mas nao eh recomendado
#dadosVendas['Imposto'] =0 
#seria a melhor opçao e menos custosa para a aplicacao
dadosVendas.loc[:,'Imposto'] = 0 
```

```python
dadosVendasDez = pd.read_excel('VendasDez.xlsx')


#renomando as colunas 
dadosVendasDez = dadosVendasDez.rename(columns={   'Código Venda':'CodigoVenda',
                                             'Data':'Data',
                                             'ID Loja':'IDLoja',
                                             'Produto':'Produto',
                                             'Quantidade':'Quantidade',
                                             'Valor Unitário':'ValorUnitario',
                                             'Valor Final':'ValorFinal'})

#mostrar apenas as 3 primeiras linhas 
dadosVendasDez.head(3)
```

```python
#unindo os DataFrames

dadosVendas = dadosVendas.append(dadosVendasDez)


dadosVendas.head(3)
```

```python
#Excluindo colunas 
# Dicas axis = 0 trata-se de linhas e axis = 0 são as colunas 
dadosVendas = dadosVendas.drop("Imposto",axis = 1)
dadosVendas.head(3)
```

```python
#excluindo linhas 
dadosVendas = dadosVendas.drop(0,axis = 0)
dadosVendas.head(3)
```

```python
#delatar linhas e colunas completamete vazias

dadosVendas = dadosVendas.dropna(how = 'all')

#detalhe caso querira deletar as colunas dadosVendas = dadosVendas.dropna(how = 'all', axis = 1)
```

```python
#deletar linhas que possuem pelo menos 1 valor vazio
dadosVendas = dadosVendas.dropna()
```

```python
#preecher valores vazios
#preencer com media da colunas
dadosVendas['Comissao'] = dadosVendas['Comissao'].fillna(dadosVendas['Comissao'].mean())
dadosVendas.head(3)
```

```python
#preencher com o ultimo valor
dadosVendas = dadosVendas.ffill()
```

```python

transacoesLoja = dadosVendas['IDLoja'].value_counts()
transacoesLoja.head(10)
```

```python

dadosVendas[['Produto','ValorFinal']].groupby('Produto').sum()
```

```python
dadosGerentes = pd.read_excel ('Gerentes.xlsx')
```

```python
dadosGerentes = dadosGerentes.rename(columns ={'ID Loja': 'IDLoja'})

dadosGerentes.head()
```

```python
#fazendo um merge de dados entre dois excel e convertendo para um DataFrame
dadosVendas = dadosVendas.merge(dadosGerentes)
dadosVendas.head(3)
```
