```python
#graficos com regressão

%matplotlib inline
sns.lmplot(x='total_bill',y='tip',hue = 'sex',palette = 'pastel', data = tips)
```

```python

%matplotlib inline
sns.lmplot(x='total_bill',y='tip',hue = 'smoker',col = 'time',palette = 'Set2', data = tips)
```

```python
sns.lmplot( x='total_bill'
           ,y='tip'
           ,hue = 'smoker'
           ,col = 'time'
           ,palette = 'Set1'
           ,data = tips
           ,scatter_kws ={'s':90,'alpha':0.5}# tamanho dos pontos
           ,line_kws = {'lw':4} # espe das linhas)
```

```python
sns.lmplot(x='total_bill',y='tip',hue = 'smoker',col = 'time',row = 'sex',palette = 'Set2', data = tips)
```
